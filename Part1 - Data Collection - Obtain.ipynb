# Import Libraries & Data Set

```python
import pandas as pd
```

```python
df = pd.read_csv('online_shoppers_intention.csv')
```

# Feature Exploration & Engineering

```python
df.info()
```

```python
df.describe()
```

```python
#change weekend (Yes/No Boolean) and Revenue/Purchase (Yes/No Boolean) to integer
df['Weekend'] = df['Weekend'].astype(int)
df['Purchase'] = df['Revenue'].astype(int)
```

```python
df['Month'].value_counts()
```

```python
#convert month to numerical data (dummies)

#change June to Jun so that all months have 3 characters and conform with "abbreviated" month names
df['Month'].replace('June', 'Jun', inplace = True)


df_month_dummies = pd.concat([df, pd.get_dummies(df['Month'].astype('category'), prefix = 'Month')], axis = 1)
df_month_dummies.drop('Month_Aug', axis=1, inplace=True) #dropping August because no holidays and small number of observations
df_month_dummies

#---ignore below (incorrect)
# #change from abbreviated month name (%b) to number of month, then convert that number to integer
# df['Month'] = pd.to_datetime(df.Month, format='%b').dt.month
# df['Month'].astype(int)
```

```python
df.describe()
```

```python
df['VisitorType'].value_counts()
```

```python
#convert VisitorType to dummy variables to be able to incorporate into logistic regression or SVM model

df_visitor_dummies = pd.concat([df_month_dummies, pd.get_dummies(df['VisitorType'].astype('category'), prefix = 'VisitorType')], axis = 1)
df_visitor_dummies.sort_values('VisitorType_Other', ascending=False)

#options: one-hot encoding; or classifiers? 
#There are different ways to treat dummies: can be done in Pandas or SKLearn; or for binary categorical variables: 1 vs. -1 (instead 1 vs. 0)--which helps regularize BOTH categories equally [instead of not regularizing the 0 variables]
#Can drop "other" since it's not interpretable anyway AND because there are so few of them 
#Nevertheless, dropping dummies is MOST needed for linear regression models because they are singular 
```

```python
df_visitor_dummies.drop('VisitorType_Other', axis=1, inplace=True)
df_visitor_dummies.describe()
#Can calculate coefficients from linreg model without vs. with dummy variable, but "Other" was a fitting choice for me so I can get coefficients in my model 
```

```python
#drop original VisitorType column (strings)
#df_dummies.drop('VisitorType',axis=1, inplace=True)
df_visitor_dummies.drop(['Revenue','Month','VisitorType'],axis=1, inplace=True)
df_visitor_dummies.info()
```

```python
df_operatingsystems_dummies = pd.concat([df_visitor_dummies, pd.get_dummies(df_visitor_dummies['OperatingSystems'].astype('category'), prefix = 'OperatingSystem')], axis = 1)
df_operatingsystems_dummies.drop('OperatingSystem_5', axis=1, inplace=True) #dropping 5 because small number of observations
df_operatingsystems_dummies
```

```python
df_browser_dummies = pd.concat([df_operatingsystems_dummies, pd.get_dummies(df_operatingsystems_dummies['Browser'].astype('category'), prefix = 'Browser')], axis = 1)
df_browser_dummies.drop('Browser_9', axis=1, inplace=True) #dropping 9 because small number of observations
df_browser_dummies
```

```python
df_region_dummies = pd.concat([df_browser_dummies, pd.get_dummies(df_browser_dummies['Region'].astype('category'), prefix = 'Region')], axis = 1)
df_region_dummies.drop('Region_5', axis=1, inplace=True) #dropping 5 because small number of observations
df_region_dummies
```

```python
df_traffictype_dummies = pd.concat([df_region_dummies, pd.get_dummies(df_region_dummies['TrafficType'].astype('category'), prefix = 'TrafficType')], axis = 1)
df_traffictype_dummies.drop('TrafficType_17', axis=1, inplace=True) #dropping 17 because small number of observations
df_traffictype_dummies
```

```python
df_traffictype_dummies.drop(['OperatingSystems','Browser','Region', 'TrafficType'],axis=1, inplace=True)
df_traffictype_dummies
```

```python
#pickle my new table!
shopping_intent_df= df_traffictype_dummies.to_pickle('shopping_intent_df')
```
