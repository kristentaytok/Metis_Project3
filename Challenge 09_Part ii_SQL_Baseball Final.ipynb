# What was the total spent on salaries by each team, each year?

````python
#LIMIT and ORDER were not specified in the question--so I did not set a limit and chose my own order
query = """
rollback;
SELECT SUM(salary), teamid, yearid 
FROM salaries
GROUP BY teamid, yearid
ORDER BY teamid, yearid;
"""
cursor.execute(query)
cursor.fetchall()
````

````python
#By removing duplicate playerIDs from both the Master and Salaries tables, we get a different mean ($9,567,810) because all the counts and distinct counts of playerIDs are now 4668.  
query = """
rollback;
SELECT AVG(sum_salary), SUM(sum_salary), COUNT(distinct_salaries.playerID), COUNT(DISTINCT(distinct_salaries.playerID)), COUNT(distinct_players.playerID), COUNT(DISTINCT(distinct_players.playerID))
FROM (SELECT playerID, SUM(salary) AS sum_salary FROM Salaries GROUP BY playerID) AS distinct_salaries
    LEFT JOIN (SELECT playerID FROM Master GROUP BY playerID) AS distinct_players
    ON distinct_salaries.playerID = distinct_players.playerID
; 
"""
cursor.execute(query)
cursor.fetchall()
````

"Salaries LEFT JOIN Master" returns an AVERAGE salary for all players that appear in the Salary and Master list (4668 distinct players). With a smaller number of players in the denominator, I would expect the MEAN SALARY of the entire list (total salary divided by number of players) becomes larger. However, the AVG(Salary) divides by only the playerIDs that the two tables share--resulting in the same MEAN SALARY.  

"Master LEFT JOIN Salaries" returns an AVERAGE salary for all players that appear in the Salary and Master list (4668 distinct players), but also 13,686 distinct players where salary is null ("None"). With a larger number of players in the denominator, I would expect the MEAN SALARY of the entire list (total salary divided by number of players) to becomes smaller. However, the AVG(Salary) divides by only the playerIDs that the two tables share--resulting in the same MEAN SALARY.   

Because the calculated average in a SQL query for two joined tables is always based on the number of observations that are present in BOTH tables, the table JOIN order does NOT affect the mean salary.
