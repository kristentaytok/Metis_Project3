# Which team scored the most points when playing at home?  

````python
#view all column names in table

cursor.execute("PRAGMA table_info(Match);")
cursor.fetchall()
````

````python
query = """
SELECT home_team_api_id, SUM(home_team_goal)
FROM Match
GROUP BY home_team_api_id
ORDER BY SUM(home_team_goal) DESC
LIMIT 1;
"""
cursor.execute(query)
cursor.fetchall()
````

```
Answer: 8633
```

# Did this team also score the most points when playing away?  

````python
query = """
SELECT away_team_api_id, SUM(away_team_goal), RANK () OVER (ORDER BY SUM(away_team_goal) DESC)
FROM Match
GROUP BY away_team_api_id;
"""
cursor.execute(query)
cursor.fetchall()
````

```
Answer:  No--they scored the second most points when playing away
```

# What was the median tie score? Use the value determined in the previous question for the number of tie games. *Hint:* PostgreSQL does not have a median function. Instead, think about the steps required to calculate a median and use the [`WITH`](https://www.postgresql.org/docs/8.4/static/queries-with.html) command to store stepwise results as a table and then operate on these results. 

````python
#Did not use... Would need to use WHEN, THEN statements for even- vs. odd-numbered lists to get a true median query
query = """
WITH tied_match_score AS 
    (SELECT match_api_id, home_team_goal AS tie_score
    FROM Match WHERE home_team_goal = away_team_goal ORDER BY tie_score)
SELECT COUNT(tie_score) % 2 FROM tied_match_score;
"""
cursor.execute(query)
cursor.fetchall()
````

````python
#Knowing the count is 6596 from Question #3, I made a subquery taking the average of the middle 2 numbers (6596/2 and the number after in the ordered list of tied scores)
query = """
WITH tied_match_score AS 
    (SELECT match_api_id, home_team_goal AS tie_score
    FROM Match WHERE home_team_goal = away_team_goal ORDER BY tie_score)
SELECT AVG(tie_score) FROM (SELECT tie_score FROM tied_match_score LIMIT 6596/2, 2);"""
cursor.execute(query)
cursor.fetchall()
````

```
Answer: 1
```

# What percentage of players prefer their left or right foot? *Hint:* Calculate either the right or left foot, whichever is easier based on how you setup the problem.

````python
#View all columns in the table
cursor.execute("PRAGMA table_info(Player_Attributes);")
cursor.fetchall()
````

````python
query = """
SELECT 100.0 * COUNT(preferred_foot='left') / (SELECT COUNT(preferred_foot) FROM Player_Attributes) AS percentage, preferred_foot
FROM Player_Attributes
GROUP BY preferred_foot;
"""
cursor.execute(query)
cursor.fetchall()
````

````python
#not working... somehow "SELECT(COUNT(preferred_foot='left'))"  results in ALL counts separated into left and right, so dividing this statement by count(preferred_foot) = 1/1 
query = """
SELECT (SELECT COUNT(preferred_foot='left') FROM Player_Attributes GROUP BY player_api_ID) / (SELECT COUNT(preferred_foot) FROM Player_Attributes GROUP BY player_api_ID) AS percentage, preferred_foot
FROM (SELECT * FROM Player_Attributes GROUP BY player_api_ID)
GROUP BY preferred_foot;
"""
cursor.execute(query)
cursor.fetchall()
````

````python
#Used a WITH operation to create a temporary table of preferred_foot per DISTINCT player_api_ID
query = """
WITH distinct_player_foot AS (SELECT preferred_foot, player_api_ID FROM Player_Attributes GROUP BY player_api_ID)
SELECT 100.0 * COUNT(preferred_foot='left')/(SELECT COUNT(preferred_foot) FROM distinct_player_foot) AS percentage, preferred_foot
    FROM distinct_player_foot
    GROUP BY preferred_foot;
"""
cursor.execute(query)
cursor.fetchall()
````

```
Answer: 24.3% prefer their left foot
```
